# Distribuiçao Amostral e Intervalo de Confiança

```python
#Armazena a coluna salario na variavel salarios
salarios = dados['SALARIO']
```

```python
salarios
```

```python
#Calcula a media de salarios e armazena na variavel media_amostral
media_amostral = np.mean(salarios)
media_amostral
```

```python
#Calcula o desvio (std) de salarios e armazena na variavel media_amostral
desvio_amostral = np.std(salarios)
desvio_amostral
```

```python
#Definindo o nivel de confiança
nivel_confianca = 0.95
```

```python
tamanho_amostra = len(salarios)
tamanho_amostra
```

```python
#Importando o stats da Biblioteca SCIPY
from scipy import stats
```

```python
#Calculando o erro padrão
erro_padrao = stats.sem(salarios)
erro_padrao
```

```python
#Calculando Intervalo de Confiança
intervalo_confianca = stats.t.interval(nivel_confianca,tamanho_amostra-1, loc=media_amostral, scale=erro_padrao)
intervalo_confianca
```

```python
#Criamos uma função chamada preencher_nivel que atribui o valor "Pessoa Gestora" se a coluna "GESTOR" for igual a 1; caso contrário, mantém o valor da coluna "NIVEL"
def preencher_nivel(gestor,nivel):
  if gestor == 1:
    return "Pessoa Gestora"
  else:
    return nivel
```

```python
dados['NOVO_NIVEL'] = dados.apply(lambda x: preencher_nivel(x['GESTOR?'], x['NIVEL']), axis=1)
```

```python
#Desenvolvemos uma função chamada determinar_geracao para categorizar os indivíduos em diferentes gerações (Geração X, Millennial, Geração Z) com base na idade.
def determinar_geracao(idade):
  if 39< idade <=58:
    return "Geração X"
  elif 29 < idade <= 39:
    return "Millenial"
  elif 13< idade <= 29:
    return "Geraçao Z"
  else: "Outra Geração"
```

```python
from scipy.stats import chi2_contingency
```

```python
def cramer_coeficiente(coluna1, coluna2):
  tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2))
  chi2 = chi2_contingency(tabela_cruzada)[0]
  soma = np.sum(tabela_cruzada)
  mini = min(tabela_cruzada.shape)-1
  cramer = np.sqrt(chi2/(soma*mini))
  return cramer
```
